```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
import pandas as pd
k=warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('COVID-19.csv')
```

```python
day=df.day
month = df.month
year=df.year
cases=df.cases
deaths=df.deaths
country=df.countriesAndTerritories
Exp = df.continentExp
```

```python
#2a---Tổng số "cases" và "deaths" theo từng ngày. (bắt đầu từ ngày 01/01/2020).
```

```python
df.head()
```

```python
df1=df.loc[(year == 2020),['day','month','year','cases','deaths']]
df1
```

```python
ngay=list(set(day))
thang=list(set(month))
quocgia=list(set(country))
```

```python
#theo ngay
for i in range(len(thang)):
    for j in range(len(ngay)):
        ngay_cases = df1.loc[(df1.day==ngay[j]) & (df1.month == thang[i])].cases.sum()
        ngay_deaths = df1.loc[(df1.day==ngay[j]) & (df1.month == thang[i])].deaths.sum()
        if(thang[i]==12):
            print("ngay ",j+1,"\t thang ",i+2,"\t cases ",ngay_cases,"\t deaths :",ngay_deaths)
        else:
            print("ngay ",j+1,"\t thang ",i+1,"\t cases ",ngay_cases,"\t deaths :",ngay_deaths)
```

```python
#theo thang
for i in range(len(thang)):
        thang_cases = df1.loc[(df1.month == thang[i])].cases.sum()
        thang_deaths = df1.loc[(df1.month == thang[i])].deaths.sum()
        if(thang[i]==12):
            print(" thang ",i+2," cases :",thang_cases," deaths :",thang_deaths)
        else:
            print(" thang ",i+1," cases ",thang_cases," deaths :",thang_deaths)
```
